**Chained Indexing**
# Dangerous

```
a = [1, 5, 10, 3, 99, 5, 8, 20, 40]

b = a[2:6][0]
b
```

```
a[2:6][0] = 55
a
```

`Nothing Happened to a`

```
a = [1, 5, 10, 3, 99, 5, 8, 20, 40]

b = a

b[0] = 55

a
```

**OMG!** a changed

```
id(a) == id(b)
```

a and b reffer to same address

**COPY function**
Solution to that problem

```
a = [1, 5, 10, 3, 99, 5, 8, 20, 40]

b = a.copy()

b[0] = 55

a
```

**Drop**

```
df.drop(0)
```

```
df
```

df is not changed. drop just returned a copy.

```
df.drop(0, inplace=True)
```

```
df
```

**Pandas Slicing Problem**

```
CENSUS2010POP = df['CENSUS2010POP']

CENSUS2010POP *= 2

df.head()
```

**df changed !!!!**

copy function is the solution

```
ESTIMATESBASE2010 = df['ESTIMATESBASE2010'].copy()

ESTIMATESBASE2010 *= 2

df.head()
```
